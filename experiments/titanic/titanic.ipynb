```python
# データを横持ちへピボットする

embarked_df = train_df[["Embarked", "Survived", "PassengerId"]].dropna().groupby(["Embarked", "Survived"]).count().unstack()
```

```python
embarked_df
```

```python
# 積み上げ縦棒グラフ

embarked_df.plot.bar(stacked=True)
```

乗船港によって、乗客のタイプや状態に何か違いがあるかもしれない。

```python
# 数値で確認

embarked_df["death_rate"]=embarked_df.iloc[:, 0] / (embarked_df.iloc[:, 1] + embarked_df.iloc[:, 0])
embarked_df
```

```python
# 性別やチケットの階級について可視化する

sex_df = train_df[["Sex", "Survived", "PassengerId"]].dropna().groupby(["Sex", "Survived"]).count().unstack()
sex_df.plot.bar(stacked=True)
```

女性の方が生存率が高い。

```python
ticket_df = train_df[["Pclass", "Survived", "PassengerId"]].dropna().groupby(["Pclass", "Survived"]).count().unstack()
ticket_df.plot.bar(stacked=True)
```

チケットの階級が高いほど、生存率が高い。

```python
train_df[train_df["Survived"] == 0][["Age"]].values
```

```python
# 年代ごとの生存率を可視化する
# 年齢は連続値のためヒストグラムを使用する。

age_dead      = train_df.loc[train_df["Survived"] == 0, "Age"].dropna()
age_survived  = train_df.loc[train_df["Survived"] == 1, "Age"].dropna()

plt.hist(
    [age_dead, age_survived], 
    histtype="barstacked",
    bins=8,
    label=("Death", "Survived")
    )

plt.legend()
```

```python
# カテゴリ変数をダミー変数化する

train_df_corr = pd.get_dummies(train_df, columns=["Sex"], drop_first=True, dtype=int)

train_df_corr = pd.get_dummies(train_df_corr, columns=["Embarked"], dtype=int)

train_df_corr.head()
```

```python
# 相関行列を作成
# 数値型の列だけを抽出
train_df_corr_selected = train_df_corr.select_dtypes(include=["number"])
train_corr = train_df_corr_selected.corr()

train_corr
```

```python
# ヒートマップで可視化する

plt.figure(figsize=(9, 9))
sns.heatmap(train_corr, vmax=1, vmin=-1, center=0, annot=True)
```

- Surviveと最も相関が高いのは「Sex_male」で-0.54。
- 次に相関が高いのは「Pclass」で-0.34。
- Fare（運賃）は0.26で相関が高い。

## 年齢は相関が低いが？

Survivedに対してAgeは相関が-0.077と低いが、年齢が生存に関係がないとは限らない。

相関係数は２変数間で、一方が上がる傾向かあるときもう一方が下がる傾向があるかを表すもの。
年齢全体が変数であるため、年齢の一部である１０代の生存率が高かったとして、全体として生存率が高くなる傾向がない場合、
相関係数は低くなるため。
